```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# URL to scrape
url = "https://africa.espn.com/football/team/squad/_/id/598/ger.fc_union_berlin"

# Set up the WebDriver (ensure you have the correct driver for your browser installed)
driver = webdriver.Chrome()
driver.get(url)

# Get the page source
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find all tables with the specific class attribute
target_tables = soup.find_all('table', class_='Table')

# Initialize a list to hold data for each table
data_tables = []

# Process each table found
for table_index, table in enumerate(target_tables):
    # Extract headers
    headers = [header.get_text(strip=True) for header in table.find_all('th')]

    # Extract data rows
    rows = table.find_all('tr')[1:]  # Skip the header row
    data = []
    for row in rows:
        cells = row.find_all('td')
        row_data = [cell.get_text(strip=True) for cell in cells]
        data.append(row_data)

    # Create DataFrame for the table
    df = pd.DataFrame(data, columns=headers)
    data_tables.append(df)

    # Save DataFrame to a CSV file
    df.to_csv(f'table_{table_index + 1}.csv', index=False)

    # Print extracted data in tabular form
    print(f"\nTable {table_index + 1}:\n")
    print(df.to_string(index=False))
```

```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup

# Set up WebDriver
driver = webdriver.Chrome()  # Ensure ChromeDriver is in your PATH or provide the executable path

# URL
url = "https://fbref.com/en/comps/8/schedule/Champions-League-Scores-and-Fixtures"

try:
    # Navigate to the page
    print(f"Navigating to URL: {url}")
    driver.get(url)
    
    # Wait for the page to load
    time.sleep(10)  # Adjust the sleep time if necessary
    
    # Parse the page
    page_soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Find all tables on the page
    tables = page_soup.find_all("table")
    
    # Print the number of tables found for verification
    print(f"Number of tables found: {len(tables)}")
    
    # Extract the first table which should be the scores and fixtures table
    if tables:
        fixtures_table = tables[0]
        
        # Extract the table headers
        headers = [header.text for header in fixtures_table.find_all("th")]
        
        # Extract the table rows
        rows = []
        for row in fixtures_table.find_all("tr"):
            rows.append([cell.text for cell in row.find_all(["td", "th"])])
        
        # Save the table data to a CSV file
        with open("Champions_League_Scores_and_Fixtures.csv", "w", encoding="utf-8") as file:
            for header in headers:
                file.write(header + ",")
            file.write("\n")
            for row in rows:
                file.write(",".join(row) + "\n")
                
        print("Table data saved successfully as 'Champions_League_Scores_and_Fixtures.csv'.")
    else:
        print("No tables found on the page.")
    
finally:
    # Close the WebDriver
    driver.quit()

print("Processing complete.")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import os
import time

# Setting up WebDriver
driver = webdriver.Chrome()


team_url = "https://www.espn.com/soccer/team/squad/_/id/359/league/UEFA.CHAMPIONS"  
team_name = "Arsenal"

# Navigate to the team squad page
driver.get(team_url)
time.sleep(3)  # Wait for the page to load

# Parse the squad page
squad_page = BeautifulSoup(driver.page_source, 'html.parser')

# Find all tables with the specific class attribute
target_tables = squad_page.find_all('table', class_='Table')

# Create a directory to save the CSV files if it doesn't exist
output_dir = 'team_squads'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process each table found
for table_index, table in enumerate(target_tables):
    # Extract headers
    headers = [header.get_text(strip=True) for header in table.find_all('th')]
    
    # Extract data rows
    rows = table.find_all('tr')[1:]  # Skip the header row
    data = []
    for row in rows:
        cells = row.find_all('td')
        row_data = [cell.get_text(strip=True) for cell in cells]
        data.append(row_data)
    
    # Create DataFrame for the table
    df = pd.DataFrame(data, columns=headers)
    
    # Determine file name based on table type
    if table_index == 1:
        table_type = 'outfield players'
    elif table_index == 0:
        table_type = 'goalkeepers'
    else:
        table_type = f'other_{table_index + 1}'
    
    # Save each DataFrame to a CSV file named after the team and table type
    output_file = os.path.join(output_dir, f"{team_name}_{table_type}_table.csv")
    print(f"Saving file to: {output_file}")
    df.to_csv(output_file, index=False)

    # Verify the file was created
    if os.path.exists(output_file):
        print(f"File saved successfully: {output_file}")
    else:
        print(f"Failed to save file: {output_file}")

# Close the WebDriver
driver.quit()

print("Processing complete.")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import os
import time

# Set up WebDriver
driver = webdriver.Chrome()

# URL for the UEFA Champions League standings
url = "https://africa.espn.com/football/table/_/league/uefa.champions"

# Navigate to the page
driver.get(url)
time.sleep(5)  # Wait for the page to load completely

# Parse the page
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')

# Close the WebDriver
driver.quit()

# Find all divs with the class 'standings_table'
standings_divs = soup.find_all('div', class_='standings_table')

# Create a directory to save the CSV files if it doesn't exist
output_dir = 'standings_data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process each standings div found
for div_index, standings_div in enumerate(standings_divs):
    # Extract group name
    group_name_tag = standings_div.find_previous('h2')
    group_name = group_name_tag.get_text(strip=True) if group_name_tag else f"Group_{div_index + 1}"

    # Find the table within the div
    table = standings_div.find('table')
    
    if not table:
        print(f"No table found for {group_name}. Skipping this div.")
        continue
    
    # Extract headers
    headers = [header.get_text(strip=True) for header in table.find_all('th')]
    if not headers:
        print(f"No headers found for {group_name}. Skipping this table.")
        continue

    # Extract data rows
    rows = table.find_all('tr')[1:]  # Skip the header row
    data = []
    for row in rows:
        cells = row.find_all('td')
        row_data = [cell.get_text(strip=True) for cell in cells]
        if len(row_data) != len(headers):
            print(f"Row data length mismatch in {group_name}. Expected {len(headers)} columns, got {len(row_data)}.")
            continue
        data.append(row_data)
    
    if not data:
        print(f"No data rows found for {group_name}. Skipping this table.")
        continue

    # Create DataFrame for the table
    try:
        df = pd.DataFrame(data, columns=headers)
    except ValueError as e:
        print(f"Error creating DataFrame for {group_name}: {e}")
        continue
    
    # Save DataFrame to CSV file named after the group
    output_file = os.path.join(output_dir, f"{group_name}.csv")
    print(f"Saving file to: {output_file}")
    df.to_csv(output_file, index=False)

```
